```python
from collections import defaultdict 

machine, human = 'x','o' # here machine move is 'x' and human move = 'o'
MIN,MAX = -1000,1000
```

```python

def evaluate(board):
   
    
 
    # Checking every coulumn
    for col in range(3) :
      
        if (board[0][col] == board[1][col] and board[1][col] == board[2][col]) :
         
            if (board[0][col] == machine) :
                return 1
            elif (board[0][col] == human) :
                return -1
    
    # checking every row
    for row in range(3) :    
        if (board[row][0] == board[row][1] and board[row][1] == board[row][2]) :       
            if (board[row][0] == machine) :
                return 1
            elif (board[row][0] == human) :
                return -1
 
    # Checking the diagonals
    if (board[0][0] == board[1][1] and board[1][1] == board[2][2]) :
     
        if (board[0][0] == machine) :
            return 1
        elif (board[0][0] == human) :
            return -1
 
    if (board[0][2] == board[1][1] and board[1][1] == board[2][0]) :
     
        if (board[0][2] == machine) :
            return 1
        elif (board[0][2] == human) :
            return -1
 
    # if the board has no winning positions, return 0
    return 0
```

```python
def check(board) :
 
    for i in range(3) :
        for j in range(3) :
            if (board[i][j] == '_') :
                return True
    return False
```

```python
def prune(board,depth,isMachine,alpha,beta):
    
    score = evaluate(board)
    
    states[tuple(map(tuple,board))]+=1
    
    if(score==1):
        return score
    if(score==-1):
        return score
    
    if(check(board)==False):
        return 0
    
    if(isMachine):
        best = MIN
        
        for i in range(3):
            for j in range(3):
                
                if(board[i][j]=='_'):
                    board[i][j] = machine
                    
                    val = prune(board,depth+1, not isMachine,alpha,beta)
                    
                    best = max(best,val)
                    alpha = max(alpha,best)
                    
                    if(beta<=alpha):
                        break
                    
                    board[i][j] = '_'
        return best
    else :
        best = MAX
        
        for i in range(3):
            for j in range(3):
                
                if(board[i][j]=='_'):
                    board[i][j] = human
                    
                    val = prune(board,depth+1,not isMachine,alpha,beta)
                    
                    best = min(best,val)
                    beta = min(beta,best)
                    
                    if(beta<=alpha):
                        break
                    
                    board[i][j] = '_'
        return best
    
        
        
                    
                    
        
    
```
